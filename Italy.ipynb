```python
import geopandas
import imageio

import matplotlib.pyplot as plt
```

```python
ruta_reg = "./"
ruta_ed = "./"
archivo_reg = "Com01012019_WGS84.shp"
archivo_ed = "gadm41_ITA_1.shp"

reg = geopandas.read_file(ruta_reg+ archivo_reg)
edo = geopandas.read_file(ruta_ed+ archivo_ed)

display(reg)
display(edo)
```

```python
edo["boundary"] = edo.boundary
edo["area"] = edo.area
edo["centroid"] = edo.centroid

reg = reg.set_index("COMUNE")
reg["boundary"] = reg.boundary
reg["area"] = reg.area
reg["centroid"] = reg.centroid
```

```python
ciudades = ["Firenze", "Verona", "Genova", "Siena", "Venezia", "Torino", "Milano", "Bologna", "Roma", "Livorno", "Sassari"]
```

```python
cities = reg.loc[reg.index.isin(ciudades)]
cities
```

```python
edo.boundary.plot()
```

```python
edo.plot(cmap='OrRd')
```

```python
edo.plot("area", legend=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 20))

edo.plot(ax=ax, color='lightblue', edgecolor='steelblue')
cities.plot(ax=ax, color ="purple")

for i in cities.index:
    ax.annotate(i, xy =  cities.geometry.centroid[i].coords[0], ha='center', fontsize = 15 )
```
